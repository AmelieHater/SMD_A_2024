Dann zur Berechnung der Funktion, die implimentiert wird: <br>
$$\int_0^{y}\frac{1}{\tau}e^{-x/\tau}\;\mathrm{d}x = \frac{1}{\tau}\left[-\tau\cdot e^{-x/\tau}\right]_{0}^{y} = -e^{-y/\tau}+1=g(y)$$
$$\implies \mathrm{ln}(-g+1) = -\frac{y}{\tau} \iff y = -\tau \cdot \mathrm{ln}(1-g)$$
